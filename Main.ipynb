```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
path='./test/img1.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title('Original Image')
plt.show()

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='Greys')
plt.title('Grayscale')
plt.show()
```

```python
# Converting to Binary Image

ret,bin_img = cv2.threshold(img,0,255,cv2.THRESH_OTSU+cv2.THRESH_BINARY)
plt.imshow(bin_img,cmap='Greys')
plt.show()
```

```python
# segmenting img to lines

def seg_lines(img):
    lines=[]
    seg_points = []
    start = end = 0
    for i in range(len(img)-1):
        if np.count_nonzero(img[i] == 0) > 0 and np.count_nonzero(img[i-1] == 0) == 0 :
            start = i
        if np.count_nonzero(img[i]==0) > 0 and np.count_nonzero(img[i+1]==0) == 0 :
            end = i
            seg_points.append([start,end])
    for i in range(len(seg_points)):
        lines.append(img[seg_points[i][0]:seg_points[i][1]])
    return lines

lines = seg_lines(bin_img)
for i in lines:
    plt.imshow(i,cmap='Greys')
    plt.show()
```
